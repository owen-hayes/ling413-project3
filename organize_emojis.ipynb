```python
import emoji
import pandas as pd
import numpy as np
```

```python
def get_emojis(text):
    emojis = emoji.analyze(text)
    return list(x[0] for x in emojis)


category_files = [
    {"file": "1.smileys.txt", "category": "Smileys"},
    {"file": "2.people.txt", "category": "People"},
    {"file": "3.animals-and-nature.txt", "category": "AnimalsAndNature"},
    {"file": "4.food.txt", "category": "Food"},
    {"file": "5.activity.txt", "category": "Activity"},
    {"file": "6.travel-and-places.txt", "category": "TravelAndPlaces"},
    {"file": "7.objects.txt", "category": "Objects"},
    {"file": "8.symbols.txt", "category": "Symbols"},
    {"file": "9.flags.txt", "category": "Flags"},
]

all_emojis_by_category = pd.DataFrame({"Emoji": [], "Category": []})

for category_file in category_files:
    file = category_file["file"]
    category_name = category_file["category"]

    this_category_df = (
        pd.read_csv(f"all-emojis/{file}", sep="|", names=["Emoji"])["Emoji"]
        .apply(lambda text: "".join(get_emojis(text)))
        .replace("", np.nan)
        .dropna()
        .reset_index(drop=True)
    )

    this_category_df = pd.DataFrame(this_category_df)
    this_category_df['Category'] = category_name

    all_emojis_by_category = pd.concat([all_emojis_by_category, this_category_df])

all_emojis_by_category

# emoji_smileys = (
#     pd.read_csv("all-emojis/1.smileys.txt", sep="|")["Emoji"]
#     .apply(lambda text: "".join(get_emojis(text)))
#     .replace("", np.nan)
#     .dropna()
#     .reset_index(drop=True)
# )

# this_category_df = pd.DataFrame(emoji_smileys)
# this_category_df["Category"] = "Smiley"
# this_category_df


# emoji_smileys = list(emoji_smileys)
# emoji_smileys[:10]
```
